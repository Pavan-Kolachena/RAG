```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

# Load your Alice in Wonderland text (you can replace this with your own text)
with open("/content/Alice in wonder land.txt") as f:
    alice_text = f.read()

# Initialize the text splitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=4000,  # Set your desired chunk size
    chunk_overlap=20,  # Set the overlap between chunks
    length_function=len,
    is_separator_regex=False,
)

# Create LangChain Document objects (for downstream tasks)
texts = text_splitter.create_documents([alice_text])

# Print the first two chunks
print(texts[0])
print(texts[1])
```

```python
#here am using character text splitter

from langchain_text_splitters import CharacterTextSplitter

# Load your Alice in Wonderland text (you can replace this with your own text)
with open("/content/Alice in wonder land.txt") as f:
    alice_text = f.read()

# Initialize the Character Text Splitter
text_splitter = CharacterTextSplitter()

# Create LangChain Document objects (for downstream tasks)
texts = text_splitter.create_documents([alice_text])

# Print the first two chunks (individual characters)
print(texts[0])
print(texts[1])
```

The basic difference between the character text splitter and recursive character text splitter is , in CTS we do splitting based on single character and the chunk size is based on the number of character where as in RCTS we do splitting based on list of characters and we will do in a hierarichal way till chunk size are small enough

Types of Hugging face tokenizer

the different types of tokenizers used in Transformers (such as those provided by Hugging Face):

Byte-Pair Encoding (BPE):
BPE is a subword tokenization method that breaks down words into smaller subword units.
It starts with individual characters and then merges the most frequent pairs of subwords.
BPE is widely used in NLP models like GPT-2 and BERT.
Example: Tokenizing “Transformers” into [“Trans”, “formers”].
Used by models like BERT1.

WordPiece:
WordPiece is another subword tokenization technique.
It also starts with individual characters but greedily merges subword units based on likelihood.
WordPiece is used by models like BERT and RoBERTa.
Example: Tokenizing “Transformers” into [“Trans”, “##form”, “##ers”].
Used by models like BERT1.

SentencePiece:
SentencePiece is a more flexible subword tokenization method.
It can handle languages with complex morphology and character-based scripts.
SentencePiece doesn’t rely on spaces or predefined rules.
Used by models like T5 and BART.
Example: Tokenizing “Transformers” into [“▁Trans”, “form”, “ers”].
Used by models like T51.
Each tokenizer has its own advantages and is chosen based on the specific requirements of the model. Remember that consistent tokenization is crucial for proper model performance!

Word Embeddings:
Use Case: Word embeddings are commonly used in natural language processing (NLP) to represent words as vectors.
Purpose: They capture semantic relationships between words. For example, the vector for “king” minus the vector for “man” is approximately equal to the vector for “queen” minus the vector for “woman.”
Application: Word embeddings enhance question-answering systems, semantic similarity tasks, and language modeling.

Graph Embeddings:
Use Case: Graph embeddings represent graph data (nodes and edges) as vectors.
Purpose: They preserve structural information from the graph in a low-dimensional space.
Application: Useful for network analysis, recommendation systems, and social network modeling.

OpenAI Embeddings:
Description: OpenAI’s text embeddings measure the relatedness of text strings. They represent text as vectors.
Common Uses:
Search: Rank search results by relevance to a query string.
Clustering: Group text strings based on similarity.
Recommendations: Recommend items with related text strings.
Anomaly Detection: Identify outliers with little relatedness.
Diversity Measurement: Analyze similarity distributions.
Classification: Classify text strings by their most similar label.
Model Options: OpenAI offers text-embedding-3-small and text-embedding-3-large models12.


Cohere Embeddings:
Description: Cohere embeddings represent the meaning of text as a list of numbers. They are used for semantic search, clustering, and classification.
Common Uses:
Semantic Search: Find similar text passages.
Classification: Categorize user feedback.
Multilingual Support: Cohere offers embed-english-v3.0 and embed-multilingual-v3.0 models for over 100 languages3.
Compression Levels: Cohere supports various compression types (e.g., float, int8, unint8) for efficient embeddings3.

HuggingFace Embeddings:
Description: HuggingFace’s SentenceTransformers framework generates embeddings for text and images.
Common Uses:
Semantic Search: Match text to text or text to images.
Recommendation Systems: Serve relevant ads.
Chatbots: Enable context-aware responses.
Open Source: HuggingFace provides open-source tools for creating embeddings4.

There are several retrieval methods used in Retrieval-Augmented Generation (RAG) systems, each with its own strengths and use cases. Here's an overview of different retrieval methods and when to use them:
1. Keyword-based retrieval:   - Method: Uses exact keyword matching or simple text similarity.   - When to use: For straightforward queries where exact matches are sufficient.   - Pros: Simple, fast, and works well for specific terminology.   - Cons: May miss semantically related content.
2. TF-IDF (Term Frequency-Inverse Document Frequency):   - Method: Weighs terms based on their frequency in a document and rarity across documents.   - When to use: For general-purpose retrieval when you want to balance term importance.   - Pros: Simple, interpretable, and works well for many applications.   - Cons: Doesn't capture semantic meaning or word order.
3. Dense vector retrieval:   - Method: Uses neural networks to encode queries and documents into dense vectors, then compares them.   - When to use: When semantic understanding is important, especially for natural language queries.   - Pros: Captures semantic meaning, good for natural language understanding.   - Cons: Computationally intensive, requires training or fine-tuning.
4. Hybrid retrieval:   - Method: Combines multiple retrieval methods (e.g., keyword + dense vector).   - When to use: To leverage strengths of different methods and improve overall retrieval quality.   - Pros: Can provide better results than single methods alone.   - Cons: More complex to implement and tune.
5. Knowledge graph-based retrieval:   - Method: Uses structured knowledge representations to find relevant information.   - When to use: When dealing with highly structured data or complex relationships.   - Pros: Good for capturing relationships and hierarchical information.   - Cons: Requires building and maintaining a knowledge graph.
6. Learned retrieval:   - Method: Uses machine learning models trained specifically for retrieval tasks.   - When to use: When you have labeled data and want to optimize for specific retrieval criteria.   - Pros: Can be highly optimized for specific use cases.   - Cons: Requires significant training data and computational resources.
Regarding using LLMs as retrieval:
Yes, LLMs can be used for retrieval, though this is a more recent and evolving approach. This method falls under the category of "generative retrieval" or "LLM-based retrieval." Here are some key points:
1. Method: The LLM generates relevant information or queries based on the input, which is then used for retrieval.
2. Types:   - Query generation: LLM generates multiple potential queries from the user's input.   - Direct content generation: LLM generates potentially relevant content directly.   - Hypothetical document retrieval: LLM imagines ideal documents that would answer the query.
3. When to use:   - For complex queries that require reasoning or reformulation.   - When traditional retrieval methods struggle with nuanced or contextual queries.   - In domains where the LLM has strong prior knowledge.
4. Pros:   - Can handle complex, natural language queries.   - Potentially more flexible and adaptable than traditional methods.   - Can leverage the LLM's broad knowledge base.
5. Cons:   - Computationally expensive.   - May introduce hallucinations or inaccuracies.   - Can be slower than traditional retrieval methods.   - Effectiveness depends heavily on the quality and domain knowledge of the LLM.
This approach is still an active area of research and development, with ongoing work to improve its efficiency and reliability.
